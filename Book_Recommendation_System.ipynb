```python
books_with_ratings = ratings.merge(books, on='ISBN')
```

```python
books_with_ratings
```

```python
books_with_ratings.shape
```

```python
num_rating = books_with_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating
```

```python
avg_rating_df = books_with_ratings.groupby('Book-Title')['Book-Rating'].agg(lambda x: x.astype(float).mean()).reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'avg_rating'}, inplace = True)
avg_rating_df
```

```python
popular_df = num_rating.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating', ascending=False)
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_rating']]
```

COLLABORATIVE FILTERING 

```python
x = books_with_ratings.groupby('User-ID').count()['Book-Rating'] > 200
High_Ratings_Users = x[x].index
```

```python
filtered_rating = books_with_ratings[books_with_ratings['User-ID'].isin(High_Ratings_Users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Message in a Bottle')
```
